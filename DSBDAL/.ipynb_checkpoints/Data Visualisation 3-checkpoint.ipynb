```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"{column} outliers:")
    if outliers.empty:
        print("No outliers found")
    else:
        print(outliers)
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for feature in features:
    data = remove_outliers(data, feature)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
sns.boxplot(data=data, x='Species', y='SepalLengthCm', ax=axes[0, 0])
axes[0, 0].set_title('Iris Sepal Length Histogram')
sns.boxplot(data=data, x='Species', y='SepalWidthCm', ax=axes[0, 1])
axes[0, 1].set_title('Iris Sepal Width Histogram')
sns.boxplot(data=data, x='Species', y='PetalLengthCm', ax=axes[1, 0])
axes[1, 0].set_title('Iris Petal Length Histogram')
sns.boxplot(data=data, x='Species', y='PetalWidthCm', ax=axes[1, 1])
axes[1, 1].set_title('Iris Petal Width Histogram')
plt.tight_layout()

plt.show()
```

```python
# Create violin plots for each feature grouped by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.violinplot(data=data, x='Species', y='SepalLengthCm', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length Distribution by Species')

sns.violinplot(data=data, x='Species', y='SepalWidthCm', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width Distribution by Species')

sns.violinplot(data=data, x='Species', y='PetalLengthCm', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length Distribution by Species')

sns.violinplot(data=data, x='Species', y='PetalWidthCm', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width Distribution by Species')

plt.tight_layout()
plt.show()
```
