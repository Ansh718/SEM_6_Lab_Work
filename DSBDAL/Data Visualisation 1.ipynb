```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
sns.set_style("whitegrid")
```

```python
titanic_data.shape
```

```python
titanic_data.dtypes
```

```python
numeric_columns = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
for col1 in numeric_columns:
    for col2 in numeric_columns:
        if col1 != col2:
            sns.pairplot(titanic_data, x_vars=col1, y_vars=col2)
            plt.title(f"Pairplot of {col1} vs {col2} in Titanic Dataset")
            plt.show()
# group = titanic_data.groupby(['class','survived'], observed=False)
# pclass_survived = group.size().unstack()
# sns.heatmap(pclass_survived,annot=True,fmt='d')
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data = titanic_data, x = 'fare', kde=True)
plt.title('Histogram of Fare')
plt.show()
```

```python
# Line plot
plt.figure(figsize=(8, 6))
sns.lineplot(data=titanic_data, x=titanic_data.index, y='fare')
plt.title('Line Plot of Fare')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic_data, x=titanic_data.index, y='fare')
plt.title('Scatter Plot of Fare')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_data, x='fare')
plt.title('Box Plot of Fare')
plt.show()
```

```python
# Point plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=titanic_data, x=titanic_data.index, y='fare')
plt.title('Point Plot of Fare')
plt.show()
```

```python
# Count plot
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x='fare')
plt.title('Count Plot of Fare')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=titanic_data, x='fare')
plt.title('Violin Plot of Fare')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=titanic_data, x='fare', estimator=sum)
plt.title('Bar Plot of Fare')
plt.show()
```

```python
# KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=titanic_data, x='fare', shade=True)
plt.title('KDE Plot of Fare')
plt.show()
```
