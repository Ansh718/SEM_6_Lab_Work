```python
data['math score'] = pd.to_numeric(data['math score'], errors='coerce')
```

```python
data['math score'].fillna(data['math score'].mean(),inplace=True)
data['reading score'].fillna(data['reading score'].mean(),inplace=True)
data['writing score'].fillna(data['writing score'].mean(),inplace=True)
```

```python
# label encoding
label_encoder = LabelEncoder()
data['gender_encoded'] = label_encoder.fit_transform(data['gender'])
data['lunch_encoded'] = label_encoder.fit_transform(data['lunch'])
data['test preparation course_encoded'] = label_encoder.fit_transform(data['test preparation course'])
data
```

```python
# one hot encoding
categorical_data = data[['gender', 'lunch', 'test preparation course']]
onehot_encoder = OneHotEncoder()
encoded_data = onehot_encoder.fit_transform(categorical_data)
print(encoded_data)
```

```python
# dummy encoding
categorical_data = data[['gender', 'lunch', 'test preparation course']]
dummy_encoded_data = pd.get_dummies(categorical_data, drop_first=True)
print(dummy_encoded_data)
```

```python
# Min-Max Scaling
# Min-Max Scaling scales the data to a fixed range, typically between 0 and 1.
numerical_columns = ['math score', 'reading score', 'writing score']
scaler = MinMaxScaler()
scaled_data = data.copy() 
scaled_data[numerical_columns] = scaler.fit_transform(scaled_data[numerical_columns])
scaled_data
```

```python
# Z-Scaling(Standardization)
# Z-score Scaling (Standardization) scales the data so that it has a mean of 0 and a standard deviation of 1.
numerical_columns = ['math score', 'reading score', 'writing score']
scaler = StandardScaler()
scaled_data = data.copy() 
scaled_data[numerical_columns] = scaler.fit_transform(scaled_data[numerical_columns])
scaled_data
```
